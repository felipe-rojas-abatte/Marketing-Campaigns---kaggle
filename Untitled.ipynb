# Preparing Data for ML model

```python
# I separated columns into numerical, categorical and output
numerical_columns = ['age','duration','campaign','pdays','previous','emp.var.rate', 'cons.price.idx','cons.conf.idx', 'euribor3m', 'nr.employed']
categorical_columns = ['job','marital','education','default','housing','loan','contact','month','day_of_week', 'poutcome']
y = ['y']
```

```python
# Transformed nominal variables 'yes' and 'no' into numerical variables 1 and 2 respetively
df_bank_clean.loc[(df_bank_clean.y == 'yes'), 'y'] = 1
df_bank_clean.loc[(df_bank_clean.y == 'no'), 'y'] = 0
```

### 1st Model: Considering only numerical features
The first aproache I will consider will be to take into account only numerical features, given that Machine Learnong models can't recognize text as input, therefore I'll reduce the number of columns of the data set to 10.

```python
X1 = df_bank_clean[numerical_columns].values
y1 = df_bank_clean[y].values.ravel()
y1 = y1.astype('int')
print('Features X1: {} \noutput y1: {}'.format(X1.shape, y1.shape))
```

```python
# Split the data into Training and Testing considerong 75% for training and 25% for testing
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.25, random_state=123)
```

```python
# Using RandomForestClassifier Model
classifier1 = RandomForestClassifier(n_estimators=50, random_state=0)
```

```python
# Training the model
classifier1.fit(X1_train, y1_train)
```

```python
## Prediction of the 1st model
y1_pred = classifier1.predict(X1_test)
```

```python
Classification_report(classifier1, y1_pred, y1_test)
```

### 2nd Model: Convert to numerical values the categorical columns using One Hot Encoder
The second aproache I will consider will be to take into account not nonly numerical features but also categorical ones. For this propouse I will vectorize the categorical columns as a one-hot numeric array. This will increase the total number of features the model will use for train.

```python
# Categorical features
XC = df_bank_clean[categorical_columns].values
```

```python
# Vectorizer One Hot Eocnder
vectorizer_ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
X2B = vectorizer_ohe.fit_transform(XC).toarray()
```

```python
# Concatenate all the features
X2 = np.concatenate((X1, X2B), axis=1)
```

```python
print('Original categorical features XC: {} \nVectorized categorical features X2B: {} \nFeatures X2: {} \noutput y: {}'.format(XC.shape, X2B.shape, X2.shape, y1.shape))
```

```python
# Split the data into Training and Testing considerong 75% for training and 25% for testing
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y1, test_size=0.25, random_state=123)
```

```python
# Using RandomForestClassifier Model
classifier2 = RandomForestClassifier(n_estimators=50, random_state=0)
```

```python
# Training the model
classifier2.fit(X2_train, y2_train)
```

```python
## Prediction of the 2nd model
y2_pred = classifier2.predict(X2_test)
```

```python
Classification_report(classifier2, y2_pred, y2_test)
```

### 3rd Model: Optimization of hyperparameters
The third aproache I will consider will be to take into account a scan of the hyper parameter of the model for maximization of the preformance.

```python
classifier2.get_params()
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
base_rf_model = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
classifier3 = RandomizedSearchCV(estimator = base_rf_model, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
```

```python
# Train the random search model
classifier3.fit(X2_train, y2_train)
```

```python
# Show the best set of hyperparameters
classifier3.best_params_
```

```python
best_random = classifier3.best_estimator_
```

```python
y2_best_pred = best_random.predict(X2_test)
```

```python
Classification_report(best_random, y2_best_pred, y2_test)
```
